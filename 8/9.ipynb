# Missing Data

```python
import numpy as np
import pandas as pd
```

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]} #np.nan means No Assigned Number
df = pd.DataFrame(d) #DataFrame makes a table of values, as opposed to df.Series which makes a table of a singular series
df
```

```python
df.dropna(axis=1) #will drop any columns with missing values
```

```python
df.dropna() #will drop any rows with missing values
```

```python
df.dropna(thresh=2) #Can specify threshold (meaning at least how many non-values does it take before it gets removed)
```

```python
df.fillna(value='FILL') #replaces non-value with something else
```

```python
df['A'].fillna(value=df['A'].mean()) #replaces nan-value from column A with the average of the values in that column (bro what is this 😭)
```



# Groupby

```python
#Groupby lets you group together rows based off a column and combine them with an aggregate function
```

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

```python
byComp = df.groupby('Company') #groups all the company columns together
byComp.mean('Sales') #finds the mean of the sales of each company
```

```python
byComp.sum('Sales') #will find the sum of all company sales
```

```python
byComp.sum('Sales').loc['FB'] #finds the sum of sales from FB specifically
```

```python
df.groupby('Company').sum('Sales').loc['FB'] #does same thing as line 115
```

```python
df.groupby('Company').count() #Counts instances of columns (like how there are two people in one company each and two sales for each company)
```

```python
df.groupby('Company').max() #For strings, it'll take whatever comes last alphabetical order; for ints, it takes the greatest value (.min() does the oppposite)
```

```python
df.groupby('Company').describe() #gives you information about each aggregated group (the percentages are part of the standard deviation stuff)
#can add .loc['GOOG'] or whateevr in the parenthesis to find a specific group
```

#Merge
* The merge functions lets you merge DataFrames together using .merge()
* Can specify 'on' keyword, but panda automatically does it w/o it



Example #1

```python
left = pd.DataFrame({'key':['K0','K1', 'K2', 'K3'],'A':['A0', 'A1','A2','A3'],'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key':['K0','K1', 'K2', 'K3'],'C':['C0', 'C1','C2','C3'],'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right,on='key')
```

Example #2

```python
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],'key2':['K0','K0','K1','K2'],'A':['A0','A1','A2','A3'],'B':['B0','B1','B2','B3']})
right = pd.DataFrame({'key1':['K0','K0','K1','K2'],'key2':['K0','K0','K1','K2'],'C':['C0','C1','C2','C3'],'D':['D0','D1','D2','D3']})
```

```python
pd.merge(left,right) #even in cases of multiple same columns, it'll automatically merge them!!
```

#Joining
* It's a method for joining two potentially differently-indexed DataFrames into a single result DataFrame

Example #1

```python
left = pd.DataFrame({'A':['A0','A1','A2'],'B':['B0','B1','B2']}, index=['K0','K1','K2'])
right = pd.DataFrame({'C':['C0','C2','C3'],'D':['D0','D2','D3']}, index=['K0','K2','K3'])
```

```python
left.join(right)
```

```python
left.join(right, how='outer') #outer means rows, inner means columns (this can be applied to all the other methods btw they just weren't working here)
```
