```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,5,11)
y=x**2
x
```

```python
y
```

```python
plt.plot(x,y) #This will plot our lists into a graph
plt.xlabel('X Label') #Labels x-axis
plt.ylabel('Y Label') #Labels y-axis
plt.title("Title") #Lables title
plt.show() #need to put this after every cell so the graph can show itself
```

```python
#subplots allow you to have multiple graphs on the same run line
plt.subplot(1,2,1) #in order in parenthesis: the number of rows, the number of columns, the plot number of reference
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.show()
```

```python
#Object-Oriented Method for Creating Graphs
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

# The position is given as a list: [left, bottom, width, height], where all values are between 0 and 1, showing how much space they take up compared to the full figure.
#Left being how far it is from the left of the white box
#Bottom being how far it is from the bottom of the white box

axes.plot(x,y) #plots values on graph and makes line of best curve/line
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
plt.show()
```

```python
fig = plt.figure() #makes the blank canvas
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)

plt.show()
```

#How To Create Subplots using OOP

```python
fig,axes=plt.subplots(nrows=1,ncols=2) #says how many numbers of rows and columns u have, like add_axes but does it automatically
#axes.plot(x,y)

for current_axes in axes:
  current_axes.plot(x,y)
plt.tight_layout() #will fix overlaps with any of the graphs
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y) #can index call these graphs
axes[0].set_title('Slope Since 2009')

axes[1].plot(y,x)
axes[1].set_title('Slope Since 1992')

plt.show()
```

#Figure Size and DPI

```python
#figsize includes tuple which is the (width,height) of the figure in inches
#dpi is the pixels per inch size
fig = plt.figure(figsize=(8,2), dpi=100)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
plt.show()
```

```python
#Can do the same thing with subplots!!!!
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()

plt.show()
```

#Plot Appearance
* To customize color, add color='' argument in parenthesis
  * Note: Could also add custom RGB values in the quotations (ex. FF8C00)
* To change linewidth, add linewidth=[int] argument, where 1 is the default
  * Shorthand for linewidth is lw
* To change transparency, add alpha=[int] argument; transparency goes from a scale of 0-1
* Linestyle is how the line looks; changing it uses linestyle=" ", where you add whatever style you want inside the quotations
  * Shorthand linestyle to ls
* To mark where the points you've plotted are, do marker='' where you add whatever the acceptable sign for a marker is (usually it's an 'o')
  * To make marker size bigger, call markersize=[int]
    * Shorthand = ms
  * To change the marker color, do markerfacecolor='' (same rules of color apply here)
  * To change the marker edge width (for some reason), do markeredgewidth=[int]
    * To specify the marker edge color, do markeredgecolor = '' (same rules for color apply)

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='red',lw=3,alpha=0.5,ls="-",marker='o',ms=20, mfc='blue',mew=10,mec='pink')

plt.show()
```

#X and Y Limit

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='black',lw=3,ls="--")

#To set x-limit, do in below format, where the first number is the lower bound and the second number is the upper bound
ax.set_xlim([0,1])

#Same thing for y-limit
ax.set_ylim([0,2])

plt.show()
```
